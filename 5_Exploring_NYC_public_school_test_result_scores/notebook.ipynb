Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

```python
#Importamos las librerías necesarias, cargamos el dataset y revisamos las primeras 5 filas.
import pandas as pd
import numpy as np

schools = pd.read_csv("C:\\Users\\ASUS\\Desktop\\Data_Science\\DataCamp\\Associate_data_scientist_in_Python\\Projects\\5_Exploring_NYC_public_school_test_result_scores\\schools.csv")

schools.head()
```

```python
#Filtramos las escuelas con almenos el 80% del puntaje máximo en la prueba de matemáticas.
best_prom_math = schools[schools['average_math']>=800*0.8].sort_values('average_math',ascending=False)
best_math_schools = best_prom_math[['school_name','average_math']]
best_math_schools
```

```python
#Creamos una columna con la suma de los puntajes de las tres pruebas y escogemos las diez escuelas con mejor puntaje.
schools['total_SAT'] = schools['average_math']+schools['average_reading']+schools['average_writing']
schools_sorted = schools.sort_values('total_SAT',ascending=False)
top_10_schools = schools_sorted.iloc[:10]
top_10_schools = top_10_schools[['school_name','total_SAT']]
top_10_schools
```

```python
#Agrupamos por distrito y calculamos el promedio y las desviación estandar del puntaje total.
district = schools.groupby('borough')['total_SAT'].agg([np.std,np.mean])
district
```

```python
#Escogemos el distrito con la desviación estandar más grande, además redondeamos a dos cifras decimales.
district = np.round(district[['std','mean']],2)
district['num_schools'] = schools['borough'].value_counts()
largest_std_dev = district.iloc[2:3]
largest_std_dev = largest_std_dev.rename(columns={'std':'std_SAT','mean':'average_SAT'})
largest_std_dev
```
