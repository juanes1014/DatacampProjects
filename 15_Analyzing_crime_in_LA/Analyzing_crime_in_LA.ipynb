```python
#Importamos las librer칤as requeridas, adem치s cargamos el dataset como crimes y vemos las primeras 5 filas.
import pandas as pd
import numpy as np
crimes = pd.read_csv("C:\\Users\\ASUS\\Desktop\\Data_Science\\DataCamp\\Associate_data_scientist_in_Python\\Projects\\15_Analyzing_crime_in_LA\\crimes.csv", parse_dates=["Date Rptd", "DATE OCC"], dtype={"TIME OCC": str})
crimes.head()
```

```python
#Creamos una columna llamada Hour en el que guardamos los dos primeros caracteres de la columna TIME OCC.
crimes['Hour'] = crimes['TIME OCC'].apply(lambda x: x[0:2] )
crimes.head()
```

```python
#Contamos cual es la hora que presenta m치s cr칤menes.
peak_crime_hour = crimes['Hour'].value_counts(sort=True)
peak_crime_hour = int(peak_crime_hour.index[0])
peak_crime_hour
```

```python
#Cambiamos el tipo de dato de la columna Hour y creamos una columna llamada Night que entre las horas de las 10pm y las 3am tenga
#un valor de 1 y sino tenga un valor de 0.
crimes['Hour'] = crimes['Hour'].astype(int)
crimes['Night'] = crimes['Hour'].apply(lambda x: 1 if (x>=22 or x<=3) else 0)
crimes.head()
```

```python
#Creamos un dataframe llamado area_count_df el cual tiene guardado de mayor a menor las 치reas con mayor n칰mero de cr칤menes
#entre las 10pm y las 3am.
area = crimes[['AREA NAME', 'Night']]
area = area[area['Night']==1]
area_count = area['AREA NAME'].value_counts(sort=True)
area_count_df = area_count.to_frame().reset_index()
area_count_df = area_count_df.rename(columns={'TOTAL':'AREA NAME', 'count':'TOTAL'})
area_count_df
```

```python
#Creamos una columna nueva llamada victim_ages con una etiqueta dependiendo del rango de edad.
def labels(age):
    """La funci칩n clasifica la edad del individuo en diferentes rangos etarios y retorna una etiqueta correspondiente"""
    if age<=17:
        return '0-17'
    elif age>17 and age<=25:
        return '18-25'
    elif age>25 and age<=34:
        return '26-34'
    elif age>34 and age<=44:
        return '35-44'
    elif age>44 and age<=54:
        return '45-54'
    elif age>54 and age<=64:
        return '55-64'
    else:
        return '65+'
    
crimes['victim_ages'] = crimes['Vict Age'].apply(labels)
crimes.head()
```

```python
#Contamos cu치l es el rango de edad el cual sufre m치s v칤ctimas. 
victim_ages = crimes['victim_ages'].value_counts()
victim_ages
```
